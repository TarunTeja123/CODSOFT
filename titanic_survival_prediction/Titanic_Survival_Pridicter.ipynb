## Handling missing Values

```python
# Age ~20% age values are empty, so the age is numeric so i am taking median age and fill it with.
df["Age"]=df["Age"].fillna(df["Age"].median())
```

```python
df["Age"].isnull().sum()
```

```python
#Embarked column have only 2 rows empty so i decided to fill with ffill or bfill.
df['Embarked']=df['Embarked'].ffill()
```

```python
df['Embarked'].isnull().sum()
```

```python
# Cabin column nearly 80% is empty values, so we have chance to drop it. But
# Instead of deleting it just get the decks( deck-A,deck-B...) Because deck and Pclass are Important for predicting Survivals.
df['Deck']=df['Cabin'].str[0]

# Now we dont need Cabin column so deleting it
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.head(1)
```

```python
# Now Deck has the same missing values . but we are gonna fill with it according to
#if pclass-1 --> [A,B,C]
#if pclass-2--> [D,E]
#if pclass-3 -->[F,G

for i,(deck,pclass) in enumerate(zip(df['Deck'],df['Pclass'])):
    if pd.isna(deck) and pclass==1:
        df.at[i,'Deck']=np.random.choice(['A','B','C'])
    elif pd.isna(deck) and pclass==2:
        df.at[i,'Deck']=np.random.choice(['D','E'])
    elif pd.isna(deck) and pclass==3:
        df.at[i,'Deck']=np.random.choice(['F','G'])
    else:
        pass
        
    
```

```python
# encoding the categorical variable sex.
sex=pd.get_dummies(df.Sex,dtype=int,drop_first=True)
sex
```

```python
# handling embarked categorical data.
embarked=pd.get_dummies(df.Embarked,dtype=int,drop_first=True)
embarked
```

```python
# handling Deck categorical column.
deck=pd.get_dummies(df.Deck,dtype=int,drop_first=True)
deck
```

#### --->Feature Engineering.

```python
# There is siblings and parent children data . merging togeather will write as family.
df["Family"]=df['SibSp']+df['Parch']
```

```python
df.head()
```

```python
# now dont need sibsp and parch so, deleting them.
df.drop(['SibSp','Parch'],axis=1,inplace=True)
```

```python
df.head(1)
```

```python
# Now adding the encoded values to main dataset.
final_df=pd.concat([df,sex,embarked,deck],axis=1)
final_df
```

```python
# last columns are not good with labels so i renamed them like this.
final_df.rename(columns={'Q':'Port_Q','S':'Port_S','B':'Deck_B','C':'Deck_C','D':'Deck_D','E':'Deck_E','F':'Deck_F','G':'Deck_G'},inplace=True)
```

```python
final_df.head(1)
```

```python
# we dont need Name,sex,Ticket,Fare,Embarked and Deck columns.so deleting them.
final_df.drop(['Name','Sex','Ticket','Fare','Embarked','Deck'],axis=1,inplace=True)
```

```python
final_df.head(3)
```

```python
y=final_df['Survived']
y
```

```python
X=final_df
X.drop("Survived",axis=1,inplace=True)
```

```python
X.info()
```

### Model Preparing and Feature Selection.

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_Scaled=scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_Scaled,y,test_size=0.3)

```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(class_weight='balanced')
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("Accuracy_score\n",accuracy_score(y_test,y_pred))
print("Confusion_matrix\n",confusion_matrix(y_test,y_pred))
print("Classification_report",classification_report(y_test,y_pred))
```

```python
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True)
plt.show()
```
